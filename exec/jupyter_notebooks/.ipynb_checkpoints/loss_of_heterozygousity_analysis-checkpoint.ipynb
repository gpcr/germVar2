### Get LOH data ready

```r
setwd('~/Documents/Project/germVar')
library(germVar2)
library(gridExtra)

dele_muts <- rbind(subset(nsSNP_muts, uid %in% subset(nsSNP_vars, rare & (MDG|CPG)& dele)$uid), LoF_muts)

## remove no ASCAT info, homozygous variants, hemizygous variants, biallele deletion
# remove duplicat event_uid
# remove BCLAF1, CASP4, PCSK7 variants, they have abnormally low NAB and TAB
dele_muts %<>% subset(., !is.na(nA)) %>%
               subset(., ! (N_hom & AB > 0.8)) %>% 
               subset(., ! (nA==0 & nB==0)) %>% 
               subset(., !(CHROM=="X" & gender=="MALE")) %>%
               subset(., !duplicated(event_uid)) %>%
               subset(., !Gene %in% c("PCSK7")) 
            
# call LOH
dele_muts$LOH <- with(dele_muts, ifelse(nA!=0 & nB!=0, "None", ifelse(AB<0.5, "VAR_lost", "REF_lost")))
dele_muts$LOH <- factor(dele_muts$LOH, levels=c("REF_lost", "VAR_lost", "None"))
table(dele_muts$LOH)
```

### Classify Genes, plot LOH by classes

```r
## categorize Genes
dele_muts$Class2 <- with(dele_muts, as.character(levels(Class)[Class]))
dele_muts$Class2[with(dele_muts, Gene %in% c("BRCA1", "BRCA2"))] <- "BRCA1/2"
dele_muts$Class2 <- factor(dele_muts$Class2, levels=c("OG", "BRCA1/2", "H-TSG", "M-TSG", "Putative\nTSG", "Other"))

# remove variants that coincide with BRCA1/2 carriers
BRCA1_pat <- subset(dele_muts, LOH!="None" & Gene=="BRCA1")$Patient
BRCA2_pat <- subset(dele_muts, LOH!="None" & Gene=="BRCA2")$Patient
dele_muts %<>% subset(., !(CHROM=="17" & Patient %in% BRCA1_pat & Gene!="BRCA1")) %>%
               subset(., !(CHROM=="13" & Patient %in% BRCA2_pat & Gene!="BRCA2"))

# tally percentage of Allele lost
LOH_by_class <- group_by(dele_muts, Class2) %>% 
                dplyr::summarise(., N_allele = length(LOH), N_loss = sum(LOH!="None"), 
                                    N_LOH = sum(LOH=="REF_lost"), frac_loss = N_loss/N_allele)
options(repr.plot.height = 3.5, repr.plot.width = 4.)
p1 <- ggplot(aes(Class2), data=subset(dele_muts, LOH!="None")) 
p1 <- p1 + geom_bar(aes(fill=LOH), position="dodge", width=0.7) + scale_y_continuous(breaks=c(0, 200, 400))
p1 <- p1 + theme_bw() + scale_fill_manual(values=c("grey20", "grey70"), guide = guide_legend(title=NULL), label=c("Normal allele lost", "Variant allele lost"))
p1 <- p1 + ylab("Number of specific\nalleles lost") + labs(x=NULL) 
p1 <- p1+ theme(panel.grid.major.x=element_blank(), legend.position=c(0.3, 0.8))
p1 <- p1 + theme(axis.text.x=element_text(angle=45, hjust=0.5, vjust=0.7, lineheight=0.7), plot.margin = unit(c(0.5,0.3,0,0.3), "cm"))
p1 <- p1 + scale_x_discrete(labels=c("OG" = "Oncogene", "BRCA1/2" ="BRCA1/2", "H-TSG"="H-TSG", 
                           "M-TSG"="M-TSG", "Putative\nTSG"="Putative\nTSG", "Other"="Other"))
p1
```

```r
options(repr.plot.height = 2., repr.plot.width = 4.)
p2 <- ggplot(LOH_by_class, aes(Class2, frac_loss)) + geom_bar(stat="identity", width=0.35, fill="darkgrey") 
p2 <- p2  + scale_y_reverse(breaks=c(0, 0.5))
p2 <- p2 + theme_bw()  + ylab("Fraction of\nallelic loss") + labs(x=NULL) 
p2 <- p2 + theme(axis.text.x=element_blank(), plot.margin = unit(c(0,0.3,0.5,0.3), "cm"))
p2 <- p2+ theme(panel.grid.major.x=element_blank())
p2 <- p2 + theme(axis.ticks = element_blank())
p2
pdf("Figures/dele_LOH_Cat.pdf", width=4, height=4.5)
grid.arrange(p1,p2,ncol=1,heights=c(0.7,0.3))
dev.off()
```

### LOH Per gene

```r
options(repr.plot.height = 5.5, repr.plot.width = 5.5)
LOH_by_gene <- group_by(dele_muts, Class, Gene) %>% 
  dplyr::summarise(., N_allele = length(LOH), N_loss = sum(LOH!="None"), N_LOH = sum(LOH=="REF_lost"),
                   frac_loss = N_loss/N_allele, frac_LOH = N_LOH/N_loss)
LOH_by_gene %<>% subset(., N_loss >=4)
LOH_by_gene$pval <- with(LOH_by_gene, mapply(function(x, y) binom.test(x, y, 0.5, alternative="greater")$p.value, N_LOH, N_loss))
LOH_by_gene$padj <- p.adjust(LOH_by_gene$pval, method="BH")

LOH_by_gene %<>% mutate(., sig = sigSymbol(pval), Gene_text = paste(Gene, sig, sep="\n"), 
                            plot_x = frac_loss, plot_y = frac_LOH-0.03, show = (pval<=0.05  | frac_LOH>=0.8 | Class %in% c("H-TSG", "M-TSG"))  )

#LOH_by_gene %<>% nudgeXY(., c("BRCA1", "BRCA2","VHL", "SDHB", "APTX", "CLTCL1", "RNASEL", "NBN", "PALB2"), "y", -0.03)
LOH_by_gene %<>% nudgeXY(., c("CEACAM7", "WRN", "RAD51C", "TNFRSF13B", "TP53", "BAP1", "NF1", "LZTR1", "MYO1A", "MSH3"), "y", 0.04)
LOH_by_gene %<>% nudgeXY(., c("ADPRH", "CLTCL1"), "x", -0.03)
LOH_by_gene %<>% nudgeXY(., c("NBN", "RAD51C", "RAD51D", "TNFRSF10B"), "x", -0.01)
LOH_by_gene %<>% nudgeXY(., c("RNASEL", "CEACAM7", "PALB2", "MSH3", "MYO1A", "APTX"), "x", 0.01)
LOH_by_gene %<>% nudgeXY(., c("ZNF180"), "x", 0.03)

p <- ggplot(data=LOH_by_gene, aes(frac_loss, frac_LOH)) 
p <- p + geom_hline(yintercept=0.5, linetype=3, color="darkgrey", alpha=0.7, size=1)
p <- p + geom_hline(yintercept=0.25, linetype=3, color="darkgrey", alpha=0.9, size=1)
p <- p + geom_hline(yintercept=0.75, linetype=3, color="darkgrey", alpha=0.7, size=1)
# manually plot several
p <- p + geom_point(fill="#e41a1c", alpha=0.7, size = 10, shape=21, color="white", data=subset(LOH_by_gene, Gene %in% c("BRCA1", "BRCA2") ), guide=F) + theme_few()
p <- p + geom_point(fill="#377eb8", alpha=0.7, size = 9, shape=21, color="white", data=subset(LOH_by_gene, Gene %in% c("ARL11")), guide=F) + theme_few()
p <- p + geom_point(fill="#377eb8", alpha=0.7, size = 9, shape=21, color="white", data=subset(LOH_by_gene, Gene %in% c("GJB2")), guide=F) + theme_few()
p <- p + geom_point(fill="#4daf4a", alpha=0.7, size = 9, shape=21, color="white", data=subset(LOH_by_gene, Gene %in% c("USP6")), guide=F) + theme_few()
p <- p + geom_point(fill="#4daf4a", alpha=0.7, size = 9, shape=21, color="white", data=subset(LOH_by_gene, Gene %in% c("FAM166A")), guide=F) + theme_few()
p <- p + geom_point(aes(size=N_loss, color=Class), alpha=0.8, data=subset(LOH_by_gene, N_loss<30 )) + theme_few()
#p <- p + scale_fill_manual(values=c( "#e41a1c", "#984ea3", "#377eb8", "#4daf4a"), guide="none") #scale_color_brewer(palette="Set1")
p <- p + scale_color_manual(values=c( "#e41a1c", "#984ea3", "#377eb8", "#4daf4a"), guide=guide_legend(title=NULL, keywidth = 1, keyheight = 1)) #scale_color_brewer(palette="Set1")
p <- p + scale_size_continuous(trans="sqrt", range=c(2, 8), guide=guide_legend(title="Number of\nallelic loss", keywidth = 0.6, keyheight = 0.6), breaks=c(4, 10, 20))
p <- p + geom_text(data=subset(LOH_by_gene, show & sig==""), aes(plot_x, plot_y, label=Gene_text), vjust=0.5, size=3.)
p <- p + geom_text(data=subset(LOH_by_gene, show & sig!=""), aes(plot_x, plot_y, label=Gene_text), hjust=0.5, size=4., lineheight=0.65)
p <- p + theme(legend.position=c(0.9,0.3), legend.text =element_text(size=rel(.8)), legend.title =element_text(size=rel(.7)))
p <- p + xlab("Fraction of allelic loss") + ylab("Fraction of normal allele lost")
p <- p + theme( axis.text = element_text(size=rel(.9)))

p
ggsave(file="Figures/dele_LOH_per_gene.pdf", width=5.5, height=5.5)
```

### Low frequency missense variants by variants 

```r
nsSNP_lf_muts <- subset(nsSNP_muts, uid %in% subset(nsSNP_vars, !rare)$uid)

## remove no ASCAT info, homozygous variants, hemizygous variants, biallele deletion
# remove duplicat event_uid
# remove BCLAF1 PRODH, they have abnormally low NAB and TAB
nsSNP_lf_muts %<>% subset(., !is.na(nA)) %>%
               subset(., ! (N_hom & AB > 0.8)) %>% 
               subset(., ! (nA==0 & nB==0)) %>% 
               subset(., !(CHROM=="X" & gender=="MALE")) %>%
               subset(., !Gene %in% c("BCLAF1", "PRODH")) 
            # remove variants that coincide with BRCA1/2 carriers
BRCA1_pat <- subset(dele_muts, LOH!="None" & Gene=="BRCA1")$Patient
BRCA2_pat <- subset(dele_muts, LOH!="None" & Gene=="BRCA2")$Patient
nsSNP_lf_muts %<>% subset(., !(CHROM=="17" & Patient %in% BRCA1_pat & Gene!="BRCA1")) %>%
               subset(., !(CHROM=="13" & Patient %in% BRCA2_pat & Gene!="BRCA2"))


# call LOH
nsSNP_lf_muts$LOH <- with(nsSNP_lf_muts, ifelse(nA!=0 & nB!=0, "None", ifelse(AB<0.5, "VAR_lost", "REF_lost")))
nsSNP_lf_muts$LOH <- factor(nsSNP_lf_muts$LOH, levels=c("REF_lost", "VAR_lost", "None"))
table(nsSNP_lf_muts$LOH)
```

```r
options(repr.plot.height = 5.5, repr.plot.width = 5.5)
nsSNP_lf_LOH <- group_by(nsSNP_lf_muts, Gene, uid) %>% 
  dplyr::summarise(., N_allele = length(LOH), N_loss = sum(LOH!="None"), N_LOH = sum(LOH=="REF_lost"),
                   frac_loss = N_loss/N_allele, frac_LOH = N_LOH/N_loss)
nsSNP_lf_LOH %<>% subset(., N_loss >=4)
nsSNP_lf_LOH$pval <- with(nsSNP_lf_LOH, mapply(function(x, y) binom.test(x, y, 0.5, alternative="greater")$p.value, N_LOH, N_loss))
nsSNP_lf_LOH %<>% plyr::join(., list_goi[c("Gene", "Class")]) %>% plyr::join(., nsSNP_vars[c("uid", "AAChange.s")])
subset(nsSNP_lf_LOH, frac_LOH>0.75)

nsSNP_lf_LOH %<>% mutate(., sig = sigSymbol(pval), var_text = paste(paste(Gene, AAChange.s, sep="/"), sig, sep=""), 
                  plot_x = frac_loss, plot_y = frac_LOH-0.01, show = frac_LOH>0.75  )

    
p <- ggplot(data=nsSNP_lf_LOH, aes(frac_loss, frac_LOH)) 
p <- p + geom_hline(yintercept=0.5, linetype=3, color="darkgrey", alpha=0.7, size=1)
p <- p + geom_hline(yintercept=0.25, linetype=3, color="darkgrey", alpha=0.9, size=1)
p <- p + geom_hline(yintercept=0.75, linetype=3, color="darkgrey", alpha=0.7, size=1)
# manually plot several
p <- p + scale_color_manual(values=c( "#ffff33","#e41a1c", "#984ea3", "#377eb8", "#4daf4a"), guide=guide_legend(title=NULL, keywidth = 1, keyheight = 1)) #scale_color_brewer(palette="Set1")
p <- p + geom_point(aes(size=N_loss, color=Class), alpha=0.8) + theme_few() + scale_y_continuous(limits=c(0,1))
p <- p + scale_size_continuous(trans="sqrt", range=c(2, 8), guide=guide_legend(title="Number of\nallelic loss", keywidth = 0.6, keyheight = 0.6), breaks=c(4, 10, 20))
p <- p + geom_text(data=subset(nsSNP_lf_LOH, show & sig!="" ), aes(plot_x, plot_y, label=var_text), size=4.)
p <- p + geom_text(data=subset(nsSNP_lf_LOH, show & sig==""), aes(plot_x, plot_y, label=var_text), vjust=0.5, size=3.)
p <- p + theme(legend.position=c(0.5,0.1), legend.direction="horizontal", legend.text =element_text(size=rel(.8)), legend.title =element_text(size=rel(.7)))
p <- p + xlab("Fraction of allelic loss") + ylab("Fraction of normal allele lost")
p <- p + theme( axis.text = element_text(size=rel(.9)))

p
```
