### combine low frequency missense and truncation variants for testing
Only autosomal variants

```r
setwd('~/Documents/Project/germVar')
library(germVar2)
# only autosomal
lf_fisher <- rbind(subset(nsSNP_vars, CHROM!="X" & !rare)[c("uid", "Gene", "AAChange", "TEAC2", "EAN", "TAC2", "AC2", "AN2", "ESP_EA_AC", "ESP_EA_AN", "X2kG_AC")], 
                 subset(LoF_vars, CHROM!="X" & !rare)[c("uid", "Gene", "AAChange", "TEAC2", "EAN", "TAC2", "AC2", "AN2", "ESP_EA_AC", "ESP_EA_AN", "X2kG_AC")] )
```

### two-sided fisher exact test, comparing TCGA vs ESP patients with European ancestry, and TCGA vs 1000G overall

```r
assocTest <-function(x, prefix="CNTR", alt="greater"){
  # x is a vector, containing case AC, AN and control AC, AN
  NAC <- x[1]
  AN <- x[2]
  CNTR_AC <- x[3]
  CNTR_AN <- x[4]
  ### perform fisher exact test
  data <- matrix(c(NAC, AN-NAC, CNTR_AC, CNTR_AN) , nrow = 2)
  ftest <- fisher.test(data, alternative = alt)
  returns<- lapply(c( ftest$p.value, ftest$estimate, ftest$conf.int[1], ftest$conf.int[2]), function(x) signif(x, 2))
  returns <- as.data.frame(returns)
  colnames(returns) <-paste(prefix, c(".f.pval", ".odds", ".conf.lo", ".conf.hi"), sep="") 
  return(returns)
}
# EA1 with ESP_fAC
lf_fisher <- cbind(lf_fisher, do.call(rbind, apply(lf_fisher[c("TEAC2", "EAN", "ESP_EA_AC", "ESP_EA_AN")], 1, function(x) assocTest(x, "ESP", "two.sided" ))))
# compare with X2kG
lf_fisher$X2kG_AN <- 5008
lf_fisher <- cbind(lf_fisher, do.call(rbind, apply(lf_fisher[c("AC2", "AN2", "X2kG_AC", "X2kG_AN")], 1, function(x) assocTest(x, "X2kG", "two.sided"))))
```

### Adjust multiple comparison, remove variants missing in either ESP or1000G

```r
# flag variants where ESP or X2kG are missing
lf_fisher$missing <- with(lf_fisher, (ESP_EA_AC==0 | X2kG_AC==0))
lf_fisher$ESP.f.padj[!lf_fisher$missing] <- p.adjust(lf_fisher$ESP.f.pval[!lf_fisher$missing], method="BH")
lf_fisher$X2kG.f.padj[!lf_fisher$missing] <- p.adjust(lf_fisher$X2kG.f.pval[!lf_fisher$missing], method="BH")
table(lf_fisher$missing)
subset(lf_fisher, ESP.f.padj<0.1 & X2kG.f.padj<0.1 & ESP.odds>1 & X2kG.odds>1)[c("Gene", "AAChange", "ESP.f.pval", "ESP.odds","ESP.conf.lo","ESP.conf.hi", "X2kG.f.pval", "X2kG.odds")]
```
